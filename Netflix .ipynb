```python
# Q11. Mean, Median, and Mode of movie ratings
mean_rating = n['rating'].mean()
median_rating = n['rating'].median()
mode_rating = n['rating'].mode()[0]

print("Mean Rating:", round(mean_rating, 2))
print("Median Rating:", round(median_rating, 2))
print("Mode Rating:", round(mode_rating, 2))
```

```python
# Q12 . What is the Average Rating per Release Year?
avg_rating_year = n.groupby('release_year')['rating'].mean()
print(avg_rating_year.tail(10))
```

```python
sns.lineplot(x=avg_rating_year.index, y=avg_rating_year.values, marker='o')
plt.title("Average Rating by Year")
plt.xlabel("Year")
plt.ylabel("Average Rating")
plt.show()
```

```python
# Q13. Which Movies are Outliers in Revenue?
Q1 = n['revenue'].quantile(0.25)
Q3 = n['revenue'].quantile(0.75)
IQR = Q3 - Q1

outliers = n[(n['revenue'] > Q3 + 1.5*IQR)]
print("Revenue Outliers:\n", outliers[['title', 'revenue']].head())
```

```python
sns.boxplot(x=n['revenue'])
plt.title("Boxplot of Revenue")
plt.show()
```

```python
# Q14. What are the 25th, 50th, and 75th Percentiles of Ratings?
percentiles = n['rating'].quantile([0.25, 0.5, 0.75])
print(percentiles)
```

```python
# Q15 Correlation Between Popularity and Vote Count
n['popularity'].corr(n['vote_count'])
sns.scatterplot(x='popularity', y='vote_count', data=n, color='orange')
plt.title("Popularity vs Vote Count")
plt.show()
```

```python
# 17. Skewness and Kurtosis of Ratings
from scipy.stats import skew, kurtosis

skew_rating = skew(n['rating'])
kurt_rating = kurtosis(n['rating'])

print("Skewness:", round(skew_rating,3))
print("Kurtosis:", round(kurt_rating,3))

```

```python
Skewness = -0.4 (slightly left-skewed)  
Kurtosis = 0.2 (normal distribution)
```

```python
sns.histplot(n['rating'], kde=True, color='purple')
plt.title("Distribution of Ratings with KDE Curve")
plt.show()
```
